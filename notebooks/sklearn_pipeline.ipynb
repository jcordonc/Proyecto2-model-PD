```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler, RobustScaler
```

```python
dataset = pd.read_csv("../data/raw/dataset_train.csv")
dataset.head()
```

```python
# Definir X (características de entrada) y Y (variables de salida)

X = dataset.drop(columns=['C3S', 'C2S', 'C3A', 'C4AF', 'FCAO', 'Unnamed: 0'])
Y = dataset[['C3S', 'C2S', 'C3A', 'C4AF', 'FCAO']]

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, shuffle=True)
```

```python
from feature_engine.encoding import OrdinalEncoder

# Crear el pipeline
pipeline = Pipeline([
    # Paso 1: Codificación ordinal para la columna categórica
    ('ordinal_encoder', OrdinalEncoder(encoding_method='arbitrary', variables=['id_eq'])),
    
    # Paso 2: Escalado de características numéricas
    ('scaler', StandardScaler())
])

# Ejecutar el pipeline
pipeline.fit(X_train)
```

```python
x_features_train = pipeline.transform(X_train)
df_features_train = pd.DataFrame(x_features_train, columns=X_train.columns)
df_features_train
```

```python
# Alinear índices (opcional: restablecer índices)
df_features_train = df_features_train.reset_index(drop=True)
Y_train = Y_train.reset_index(drop=True)

# Combinar ambos DataFrames
df_features_train = pd.concat([df_features_train, Y_train], axis=1)
df_features_train
```

```python
# Alinear índices (opcional: restablecer índices)
X_test = X_test.reset_index(drop=True)
Y_test = Y_test.reset_index(drop=True)

# Combinar ambos DataFrames
dataset_test = pd.concat([X_test, Y_test], axis=1)
dataset_test
```

```python
# guardamos los datos para entrenar y evaluar modelos
df_features_train.to_csv('../data/processed/df_features_train.csv', index=False)
dataset_test.to_csv('../data/processed/dataset_test.csv', index=False)
```
